```python
from google import genai
import gradio as gr

# ----------------------------
# Gemini API
# ----------------------------
client = genai.Client(
    api_key="AQ.Ab8RN6IryvDZa6qv_DCpB2CW-054dvHkQX63hNHuyiPkOV7qXA"
)

# ----------------------------
# Health Chatbot Function
# ----------------------------
def health_chat(message, history):

    prompt = f"""
You are an AI Health Assistant.

Rules:
- Answer ONLY health-related questions.
- If the user asks anything else reply:
'I am a Health Chatbot. Please ask only health-related questions.'
- Give short and easy answers.
- End every answer with:
'⚠️ This information is for educational purposes only. Please consult a doctor.'

Question:
{message}
"""

    try:
        response = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=prompt
        )
        return response.text

    except Exception as e:
        return f"Error: {e}"


# ----------------------------
# Theme
# ----------------------------

theme = gr.themes.Soft(
    primary_hue="green",
    secondary_hue="emerald",
    neutral_hue="slate"
)

# ----------------------------
# Interface
# ----------------------------

with gr.Blocks(theme=theme,title="Health Chatbot") as demo:

    gr.Markdown(
        """
# 🩺 AI Health Chatbot

### 💚 Your Smart Health Assistant

Developed by **Raheela Nisar**
"""
    )

    with gr.Row():

        with gr.Column(scale=1):

            gr.Markdown("## 📋 Health Tips")

            gr.Markdown("""
✅ Drink 8 glasses of water

🥗 Eat healthy food

🏃 Exercise daily

😴 Sleep 7–8 hours

🍎 Eat fruits every day
""")

            gr.Markdown("## 💡 Example Questions")

            gr.Markdown("""
• What are symptoms of diabetes?

• What foods contain protein?

• How to reduce stress?

• How much water should I drink?

• What is blood pressure?
""")

        with gr.Column(scale=3):

            chatbot = gr.ChatInterface(
                fn=health_chat,
                chatbot=gr.Chatbot(height=500),
                textbox=gr.Textbox(
                    placeholder="Ask any health-related question...",
                    container=False
                )
            )

demo.launch()
```
